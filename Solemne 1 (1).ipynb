# Desafio 1

Para este desafio lo mejor fue considerar diferentes tipos de matrices con elementos aleatorios de -10 a 10 de cualquier tamaño, pero con unas excepciones.

Consideramos por convención que una matriz 0x0 su determinante es 1, entonces por eso no entra dentro de las posibilidades.

También el caso base 1x1 ya que cae en un bucle porque genera una matriz vacía como el caso 0x0.

```python
# Solicitar matriz n x n - solo permite números positivos
n = 0
while n <= 0:
    n = int(input("Ingresa el tamaño de la matriz (n x n): "))
    if n <= 0:
        print("Error: El tamaño debe ser un número positivo mayor que 0")
```

```python
# Crear matriz n x n con números aleatorios entre -10 y 10
matriz = []
for i in range(n):
    fila = []
    for j in range(n):
        fila.append(random.randint(-10, 10))
    matriz.append(fila)

print("Matriz generada:")
for fila in matriz:
    print(fila)
```

```python
# Resolución por algoritmo recursivo
def determinante_recursivo(matriz):
    n = len(matriz)

    #Caso base
    if n == 1:
        return matriz[0][0]
        
    det = 0
    for j in range(n):
        # Crear submatriz eliminando fila 0 y columna j
        submatriz = []
        for i in range(1, n):
            fila = [matriz[i][k] for k in range(n) if k != j]
            submatriz.append(fila)
        
        signo = 1 if j % 2 == 0 else -1
        cofactor = signo * matriz[0][j] * determinante_recursivo(submatriz) 
        det += cofactor
    
    return det

# Mostrar resultado
det_recursivo = determinante_recursivo(matriz)
print(f"Determinante con metodo recursivo: {det_recursivo}")
```

```python
# Metodo Iterativo
def determinante_iterado(matriz):
    # Copia para no modificar la original
    mat = [fila[:] for fila in matriz]
    n = len(mat)
    det = 1

    for i in range(n):
        # Buscar pivote
        pivote_fila = max(range(i, n), key=lambda j: abs(mat[j][i]))
        
        # Si el pivote es cero, determinante = 0
        if abs(mat[pivote_fila][i]) < 1e-10:
            return 0
        
        # Intercambiar filas si es necesario
        if pivote_fila != i:
            mat[i], mat[pivote_fila] = mat[pivote_fila], mat[i]
            det *= -1
        
        pivote = mat[i][i]
        det *= pivote
        
        # Eliminar elementos debajo del pivote
        for j in range(i + 1, n):
            factor = mat[j][i] / pivote
            for k in range(i + 1, n):
                mat[j][k] -= factor * mat[i][k]
    return det

# Probar el algoritmo iterativo
det_iterado = determinante_iterado(matriz)
print(f"Determinante con método iterativo: {det_iterado}")
```

```python
# Celda 1: Generar problema de asignación de memoria
import random

print("=== PROBLEMA DE ASIGNACIÓN ÓPTIMA DE MEMORIA ===")

# Parámetros del problema
n_procesos = 5
memoria_total = 50  # MB
procesos = []

# Generar procesos aleatorios
for i in range(n_procesos):
    memoria = random.randint(5, 25)
    prioridad = random.randint(1, 10)
    procesos.append((f"P{i+1}", memoria, prioridad))

# Generar restricciones de compatibilidad (algunos procesos no pueden coexistir)
restricciones = []
for i in range(n_procesos // 2):
    p1 = random.randint(0, n_procesos-1)
    p2 = random.randint(0, n_procesos-1)
    if p1 != p2 and (p2, p1) not in restricciones:
        restricciones.append((p1, p2))

print(f"Memoria disponible: {memoria_total} MB")
print(f"\nProcesos disponibles:")
for i, (nombre, mem, prio) in enumerate(procesos):
    print(f"  {nombre}: {mem} MB, prioridad {prio}")

print(f"\nRestricciones de compatibilidad:")
for p1, p2 in restricciones:
    print(f"  {procesos[p1][0]} y {procesos[p2][0]} no pueden ejecutarse juntos")
```

```python
# Celda 2: Backtracking
def asignacion_backtracking(procesos, memoria_total, restricciones):
    """
    Encuentra la asignación óptima usando backtracking
    """
    n = len(procesos)
    mejor_asignacion = []
    mejor_prioridad = 0
    mejor_memoria = 0
    
    def backtrack(indice, asignacion_actual, memoria_actual, prioridad_actual):
        nonlocal mejor_asignacion, mejor_prioridad, mejor_memoria
        
        # Si llegamos al final, verificar si es mejor solución
        if indice == n:
            if prioridad_actual > mejor_prioridad:
                mejor_asignacion = asignacion_actual.copy()
                mejor_prioridad = prioridad_actual
                mejor_memoria = memoria_actual
            return
        
        nombre, mem, prio = procesos[indice]
        
        # Opción 1: NO asignar el proceso actual
        backtrack(indice + 1, asignacion_actual, memoria_actual, prioridad_actual)
        
        # Opción 2: Asignar el proceso actual (si es válido)
        nueva_memoria = memoria_actual + mem
        if nueva_memoria <= memoria_total:
            # Verificar restricciones de compatibilidad
            valido = True
            for p1, p2 in restricciones:
                if indice == p1 and p2 in asignacion_actual:
                    valido = False
                    break
                if indice == p2 and p1 in asignacion_actual:
                    valido = False
                    break
            
            if valido:
                asignacion_actual.append(indice)
                backtrack(indice + 1, asignacion_actual, nueva_memoria, prioridad_actual + prio)
                asignacion_actual.pop()  # Backtrack!
    
    backtrack(0, [], 0, 0)
    
    # Convertir índices a nombres de procesos
    asignacion_nombres = [procesos[i][0] for i in mejor_asignacion]
    return asignacion_nombres, mejor_prioridad, mejor_memoria

print("=== ALGORITMO BACKTRACKING ===")
asign_bt, prioridad_bt, memoria_bt = asignacion_backtracking(procesos, memoria_total, restricciones)
print(f"Asignación óptima: {asign_bt}")
print(f"Prioridad total: {prioridad_bt}")
print(f"Memoria utilizada: {memoria_bt}/{memoria_total} MB")

# Mostrar detalles
print("\nDetalle de la asignación:")
for nombre in asign_bt:
    for n, mem, prio in procesos:
        if n == nombre:
            print(f"  {nombre}: {mem} MB, prioridad {prio}")
            break
```

```python
# Celda 3: Búsqueda en Profundidad (DFS)
def asignacion_dfs(procesos, memoria_total, restricciones):
    """
    Encuentra la asignación óptima usando DFS
    """
    n = len(procesos)
    mejor_asignacion = []
    mejor_prioridad = 0
    mejor_memoria = 0
    nodos_explorados = 0
    
    def dfs(indice, asignacion_actual, memoria_actual, prioridad_actual):
        nonlocal mejor_asignacion, mejor_prioridad, mejor_memoria, nodos_explorados
        
        nodos_explorados += 1
        
        # Siempre evaluamos en cada nodo (a diferencia de backtracking que poda más)
        if prioridad_actual > mejor_prioridad and memoria_actual <= memoria_total:
            # Verificar restricciones
            valido = True
            for p1, p2 in restricciones:
                if p1 in asignacion_actual and p2 in asignacion_actual:
                    valido = False
                    break
            
            if valido:
                mejor_asignacion = asignacion_actual.copy()
                mejor_prioridad = prioridad_actual
                mejor_memoria = memoria_actual
        
        # Si llegamos al final, retroceder
        if indice == n:
            return
        
        # Explorar en profundidad: NO asignar siguiente proceso
        dfs(indice + 1, asignacion_actual, memoria_actual, prioridad_actual)
        
        # Explorar en profundidad: ASIGNAR siguiente proceso
        nombre, mem, prio = procesos[indice]
        nueva_memoria = memoria_actual + mem
        if nueva_memoria <= memoria_total:
            nueva_asignacion = asignacion_actual + [indice]
            dfs(indice + 1, nueva_asignacion, nueva_memoria, prioridad_actual + prio)
    
    dfs(0, [], 0, 0)
    
    # Convertir índices a nombres
    asignacion_nombres = [procesos[i][0] for i in mejor_asignacion]
    return asignacion_nombres, mejor_prioridad, mejor_memoria, nodos_explorados

print("\n=== ALGORITMO BÚSQUEDA EN PROFUNDIDAD (DFS) ===")
asign_dfs, prioridad_dfs, memoria_dfs, nodos = asignacion_dfs(procesos, memoria_total, restricciones)
print(f"Asignación óptima: {asign_dfs}")
print(f"Prioridad total: {prioridad_dfs}")
print(f"Memoria utilizada: {memoria_dfs}/{memoria_total} MB")
print(f"Nodos explorados: {nodos}")

# Mostrar detalles
print("\nDetalle de la asignación:")
for nombre in asign_dfs:
    for n, mem, prio in procesos:
        if n == nombre:
            print(f"  {nombre}: {mem} MB, prioridad {prio}")
            break
```

```python
# Celda 4: Fuerza Bruta con Ciclos For Anidados
def asignacion_fuerza_bruta(procesos, memoria_total, restricciones):
    """
    Encuentra la asignación óptima probando todas las combinaciones
    """
    n = len(procesos)
    mejor_asignacion = []
    mejor_prioridad = 0
    mejor_memoria = 0
    combinaciones_probadas = 0
    
    # Probar todas las 2^n combinaciones posibles
    for mascara in range(1 << n):  # 2^n combinaciones
        combinaciones_probadas += 1
        
        memoria_actual = 0
        prioridad_actual = 0
        asignacion_actual = []
        
        # Construir la combinación actual
        for i in range(n):
            if (mascara >> i) & 1:  # Si el bit i está activo
                nombre, mem, prio = procesos[i]
                memoria_actual += mem
                prioridad_actual += prio
                asignacion_actual.append(i)
        
        # Verificar si es válida y mejor
        if memoria_actual <= memoria_total:
            # Verificar restricciones
            valido = True
            for p1, p2 in restricciones:
                if p1 in asignacion_actual and p2 in asignacion_actual:
                    valido = False
                    break
            
            if valido and prioridad_actual > mejor_prioridad:
                mejor_asignacion = asignacion_actual.copy()
                mejor_prioridad = prioridad_actual
                mejor_memoria = memoria_actual
    
    # Convertir índices a nombres
    asignacion_nombres = [procesos[i][0] for i in mejor_asignacion]
    return asignacion_nombres, mejor_prioridad, mejor_memoria, combinaciones_probadas

print("\n=== ALGORITMO FUERZA BRUTA ===")
asign_fb, prioridad_fb, memoria_fb, comb_probadas = asignacion_fuerza_bruta(procesos, memoria_total, restricciones)
print(f"Asignación óptima: {asign_fb}")
print(f"Prioridad total: {prioridad_fb}")
print(f"Memoria utilizada: {memoria_fb}/{memoria_total} MB")
print(f"Combinaciones probadas: {comb_probadas}")

# Mostrar detalles
print("\nDetalle de la asignación:")
for nombre in asign_fb:
    for n, mem, prio in procesos:
        if n == nombre:
            print(f"  {nombre}: {mem} MB, prioridad {prio}")
            break
```

```python
# Celda 1: Crear tablero de Sudoku
def crear_tablero_sudoku():
    """
    Crea un tablero de Sudoku 9x9 con algunos números predefinidos
    """
    # Tablero vacío (0 = casilla vacía)
    tablero = [
        [5, 3, 0, 0, 7, 0, 0, 0, 0],
        [6, 0, 0, 1, 9, 5, 0, 0, 0],
        [0, 9, 8, 0, 0, 0, 0, 6, 0],
        [8, 0, 0, 0, 6, 0, 0, 0, 3],
        [4, 0, 0, 8, 0, 3, 0, 0, 1],
        [7, 0, 0, 0, 2, 0, 0, 0, 6],
        [0, 6, 0, 0, 0, 0, 2, 8, 0],
        [0, 0, 0, 4, 1, 9, 0, 0, 5],
        [0, 0, 0, 0, 8, 0, 0, 7, 9]
    ]
    return tablero

def mostrar_tablero(tablero):
    """
    Muestra el tablero de Sudoku de forma legible
    """
    print("┌─────────┬─────────┬─────────┐")
    for i in range(9):
        if i % 3 == 0 and i != 0:
            print("├─────────┼─────────┼─────────┤")
        
        print("│", end="")
        for j in range(9):
            if j % 3 == 0 and j != 0:
                print(" │", end="")
            if tablero[i][j] == 0:
                print(" .", end="")
            else:
                print(f" {tablero[i][j]}", end="")
        print(" │")
    print("└─────────┴─────────┴─────────┘")

# Crear y mostrar el tablero inicial
print("=== SUDOKU - ALGORITMO BACKTRACKING ===")
print("\nTABLERO INICIAL:")
tablero = crear_tablero_sudoku()
mostrar_tablero(tablero)
```

```python
# Celda 2: Funciones de verificación
def es_valido(tablero, fila, columna, numero):
    """
    Verifica si es válido colocar 'numero' en la posición (fila, columna)
    """
    # Verificar fila
    for j in range(9):
        if tablero[fila][j] == numero:
            return False
    
    # Verificar columna
    for i in range(9):
        if tablero[i][columna] == numero:
            return False
    
    # Verificar subcuadrícula 3x3
    inicio_fila = (fila // 3) * 3
    inicio_columna = (columna // 3) * 3
    
    for i in range(3):
        for j in range(3):
            if tablero[inicio_fila + i][inicio_columna + j] == numero:
                return False
    
    return True

def encontrar_casilla_vacia(tablero):
    """
    Encuentra la próxima casilla vacía (valor 0)
    Retorna (fila, columna) o None si no hay casillas vacías
    """
    for i in range(9):
        for j in range(9):
            if tablero[i][j] == 0:
                return (i, j)
    return None

# Probar las funciones de verificación
print("\n=== PRUEBA DE VERIFICACIÓN ===")
print(f"¿Es válido poner 5 en (0,2)? {es_valido(tablero, 0, 2, 5)}")
print(f"¿Es válido poner 1 en (0,2)? {es_valido(tablero, 0, 2, 1)}")

casilla_vacia = encontrar_casilla_vacia(tablero)
print(f"Primera casilla vacía: {casilla_vacia}")
```

```python
# Celda 3: Algoritmo Backtracking
def resolver_sudoku(tablero):
    """
    Resuelve el Sudoku usando backtracking
    Retorna True si encontró solución, False si no hay solución
    """
    # Encontrar próxima casilla vacía
    vacio = encontrar_casilla_vacia(tablero)
    
    # Si no hay casillas vacías, el Sudoku está resuelto
    if not vacio:
        return True
    
    fila, columna = vacio
    
    # Probar números del 1 al 9
    for numero in range(1, 10):
        if es_valido(tablero, fila, columna, numero):
            # Hacer el movimiento
            tablero[fila][columna] = numero
            
            # Mostrar progreso (opcional)
            # print(f"Probando {numero} en ({fila}, {columna})")
            
            # Llamada recursiva (BACKTRACKING)
            if resolver_sudoku(tablero):
                return True
            
            # Si llegamos aquí, el movimiento no llevó a solución
            # BACKTRACK: deshacer el movimiento
            tablero[fila][columna] = 0
    
    # Si ningún número funciona, no hay solución
    return False

# Hacer una copia del tablero original para resolver
tablero_resolver = [fila[:] for fila in tablero]

print("\n=== RESOLVIENDO SUDOKU ===")
print("Calculando solución...")

if resolver_sudoku(tablero_resolver):
    print("✅ ¡Sudoku resuelto!")
    print("\nTABLERO RESUELTO:")
    mostrar_tablero(tablero_resolver)
else:
    print("❌ No se encontró solución para este Sudoku")
```

```python
# Celda 4: Backtracking con visualización
def resolver_sudoku_con_pasos(tablero, mostrar_pasos=False):
    """
    Resuelve el Sudoku mostrando el proceso paso a paso
    """
    pasos = [0]  # Contador de pasos (usamos lista para pasar por referencia)
    
    def backtrack(tablero_actual):
        pasos[0] += 1
        
        if mostrar_pasos and pasos[0] % 1000 == 0:
            print(f"Paso {pasos[0]}...")
        
        vacio = encontrar_casilla_vacia(tablero_actual)
        if not vacio:
            return True
        
        fila, columna = vacio
        
        for numero in range(1, 10):
            if es_valido(tablero_actual, fila, columna, numero):
                tablero_actual[fila][columna] = numero
                
                if backtrack(tablero_actual):
                    return True
                
                tablero_actual[fila][columna] = 0
        
        return False
    
    resultado = backtrack(tablero)
    return resultado, pasos[0]

# Resolver mostrando el proceso
print("\n=== RESOLUCIÓN CON CONTADOR DE PASOS ===")
tablero_con_pasos = [fila[:] for fila in tablero]

exito, total_pasos = resolver_sudoku_con_pasos(tablero_con_pasos, mostrar_pasos=True)

if exito:
    print(f"✅ Resuelto en {total_pasos} pasos")
    print("\nTABLERO FINAL:")
    mostrar_tablero(tablero_con_pasos)
else:
    print(f"❌ No resuelto después de {total_pasos} pasos")
```
